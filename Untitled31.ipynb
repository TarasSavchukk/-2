```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 1. Генерація випадкового набору даних (1000 значень)
np.random.seed(42) # Фіксуємо для відтворюваності
X = np.random.uniform(low=0, high=10, size=1000).reshape(-1, 1)

# Y генерується як квадратична функція від X з додаванням шуму
Y = 2 * X**2 + 5 * X + 10 + np.random.normal(0, 10, size=1000).reshape(-1, 1)
Y = Y.flatten() # Перетворюємо в одномірний масив

print(f"Розмір набору даних X: {X.shape}")
print("-" * 30)

# 2. Нормалізація значень X (StandardScaler)
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

# 3. Розділення на навчальну (train) і тестову (test) вибірки
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, Y, test_size=0.2, random_state=42
)

print(f"Розмір навчальної вибірки: {X_train.shape}")
print(f"Розмір тестової вибірки: {X_test.shape}")
print("-" * 30)
```

```python
# 4. Навчити KNN-регресор з різними значеннями К
# 5. Вибрати величину К для найкращих показників якості регресії

K_values = range(1, 41)
mse_scores = []
r2_scores = []

for k in K_values:
    # Створення, навчання та прогноз
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(X_train, Y_train)
    Y_pred = knn_regressor.predict(X_test)
    
    # Оцінка якості
    mse = mean_squared_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)
    
    mse_scores.append(mse)
    r2_scores.append(r2)

# Знаходимо оптимальне K:
# - Для R²: шукаємо максимальне значення (найкраща модель)
best_k_r2 = K_values[np.argmax(r2_scores)]
best_r2 = np.max(r2_scores)
# - Для MSE: шукаємо мінімальне значення (найменша похибка)
best_k_mse = K_values[np.argmin(mse_scores)]
best_mse = np.min(mse_scores)

print(f"Оптимальне K (за R-квадратом): K = {best_k_r2} (R^2 = {best_r2:.4f})")
print(f"Оптимальне K (за MSE): K = {best_k_mse} (MSE = {best_mse:.2f})")
print("-" * 30)
```

```python
# 6. Здійснити візуалізації отриманих рішень

# Візуалізація 1: Залежність метрики від K
plt.figure(figsize=(12, 5))

# Графік R^2
plt.subplot(1, 2, 1)
plt.plot(K_values, r2_scores, marker='o', linestyle='-', color='blue')
plt.title(f'Залежність R² від K (Оптимальне K={best_k_r2})')
plt.xlabel('Кількість сусідів (K)')
plt.ylabel('R²')
plt.axvline(best_k_r2, color='red', linestyle='--', label=f'Best K={best_k_r2}')
plt.grid(True)
plt.legend()

# Графік MSE
plt.subplot(1, 2, 2)
plt.plot(K_values, mse_scores, marker='o', linestyle='-', color='green')
plt.title(f'Залежність MSE від K (Оптимальне K={best_k_mse})')
plt.xlabel('Кількість сусідів (K)')
plt.ylabel('MSE')
plt.axvline(best_k_mse, color='red', linestyle='--', label=f'Best K={best_k_mse}')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Візуалізація 2: Прогноз моделі з оптимальним K на тестовій вибірці
optimal_k = best_k_r2 # Використовуємо K, оптимізоване за R²

# Перенавчаємо модель з оптимальним K
knn_optimal = KNeighborsRegressor(n_neighbors=optimal_k)
knn_optimal.fit(X_train, Y_train)
Y_pred_optimal = knn_optimal.predict(X_test)

# Повертаємо X_test до оригінального масштабу для коректної візуалізації
X_test_original = scaler_X.inverse_transform(X_test)
# Сортуємо дані для гладкої лінії прогнозу
sort_idx = X_test_original.flatten().argsort()
X_test_sorted = X_test_original[sort_idx]
Y_test_sorted = Y_test[sort_idx]
Y_pred_optimal_sorted = Y_pred_optimal[sort_idx]

plt.figure(figsize=(10, 6))
# Фактичні точки тестової вибірки
plt.scatter(X_test_original, Y_test, color='gray', label='Фактичні дані (тест)', alpha=0.6)
# Лінія прогнозу KNN-регресора
plt.plot(X_test_sorted, Y_pred_optimal_sorted, color='red', linewidth=2, label=f'Прогноз KNN (K={optimal_k})')

plt.title(f'Прогноз KNN Регресора vs Фактичні дані (Оптимальне K={optimal_k})')
plt.xlabel('X (Вхідна ознака)')
plt.ylabel('Y (Цільова змінна)')
plt.legend()
plt.grid(True)
plt.show()

print("Завдання виконано. Обидві візуалізації створені.")
```
